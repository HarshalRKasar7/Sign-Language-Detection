# Organizing Dataset

```python
import os
directory = r'.\data'
```

## Create Directory for each Class

```python
# Iterate over all files in the directory
for filename in os.listdir(directory):
    period_index = filename.find('.')
    if filename[period_index-1].isdigit():
        newfolder = filename[:period_index-1]
    else:
        newfolder = filename[:period_index]
    # If a period is found, return the substring before it
    if not os.path.exists(directory + '/' + newfolder):
        os.makedirs(directory + '/' + newfolder)
    os.rename(directory + '/' + filename, directory+ '/' + newfolder + '/' + filename)
```

## Remove useless xml files

```python
for sub_dir in os.listdir(directory):
    for file in os.listdir(os.path.join(directory, sub_dir)):
        if file.endswith('.xml'):
            print(file)
            os.remove(os.path.join(directory, sub_dir, file))
```

## Counting .jpg file in Directory

```python
count = 0
for sub_dir in os.listdir(directory):
    for file in os.listdir(os.path.join(directory, sub_dir)):
        count += 1
count
```
